## Data description
#### The contract table (data on contracts):
    customerID - unique user identifier
    BeginDate - beginning date of user using service        
    EndDate - end date of user using service           
    Type - type of service how often payments are made.
    PaperlessBilling - monthly statement recieved with or without paper bills.
    PaymentMethod - prefered method of payment
    MonthlyCharges - avg monthly charges 
    TotalCharges - total charges
#### The internet table (data on internet service plans):
    customerID - unique user identifier
    InternetService - internet service type
    OnlineSecurity - whether or not their internet service is protected
    OnlineBackup - whether or not their internet service is backed up/saved
    DeviceProtection - screen repair, battery replacement & same day replacement of your damaged, lost & stolen item
    TechSupport - online or phone support for billing, plans and tech issues
    StreamingTV - including a streaming service for live tv
    StreamingMovies - including a streaming service for movies
#### The personal table (data on personal info):
    customerID - unique user identifier
    gender - user gender
    SeniorCitizen - if the user is a senior citizen or not
    Partner - if the user has a partner
    Dependents - if the user has dependents
